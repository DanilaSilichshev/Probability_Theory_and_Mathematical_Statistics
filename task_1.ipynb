1. Из колоды в 52 карты извлекаются случайным образом 4 карты.

a) Найти вероятность того, что все карты – крести. <br />
Воспользуемся условной вероятностью:

$$ P = \frac{13}{52}*\frac{12}{51}*\frac{11}{50}*\frac{10}{49} \approx 0.002641$$

б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз. <br />
Общее число исходов:

$$ C^4_{52} = \frac{52!}{4!*(52-4)!} = 270725$$

Число исходов, при которых среди 4-х карт нет туза:

$$ C^4_{48} = \frac{48!}{4!*(48-4)!} = 194580$$

Вероятность того, что среди 4-х карт окажется хотя бы один туз:

$$ P = \frac{C^4_{52} - C^4_{48}}{C^4_{52}} = \frac{270725 - 194580}{270725} = \frac{76145}{270725} \approx 0.281263$$


2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

Количество сочетаний из трёх цифр от 0 до 9:

$$ C^3_{10} = \frac{10!}{3!*(10-3)!} = 120$$

Вероятность того, что человек, не знающий код, откроет дверь с первой попытки:

$$ P = \frac{1}{120} \approx 0.008333$$

3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

$$ P=\frac{C^3_9}{C^3_{15}}=\frac{\frac{9!}{3!*(9-3)!}}{\frac{15!}{3!*(15-3)!}} = \frac{\frac{9!}{3!*6!}}{\frac{15!}{3!*12!}} = \frac{\frac{7*8*9}{2*3}}{\frac{13*14*15}{2*3}} = \frac{7*8*9}{13*14*15} = \frac{12}{65} \approx 0.184615$$

4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

Воспользуемся условной вероятностью:

$$ P = \frac{2}{100}*\frac{1}{99} \approx 0.000202$$
